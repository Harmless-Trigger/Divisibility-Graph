```python
def divisibility_graph(integers):
    # Create a new graph
    G = Graph()
    
    # Add vertices for each integer
    for i in integers:
        G.add_vertex(i)
    
    # Add edges based on divisibility
    for i in integers:
        for j in integers:
            if i != j and (i % j == 0 or j % i == 0):
                G.add_edge(i, j)
    
    return G
```

```python
# Example usage
integers = [2,4,6,8,10,20,40,16]
G = divisibility_graph(integers)

# Display the graph
G.show()
```
